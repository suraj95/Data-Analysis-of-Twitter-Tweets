```python
Users=df['user'] # Collection of Place Json Objects
friends_count=[]
print(Users[0])
for user in Users:
    try:
        friends_count.append(user['friends_count'])
    except:
        pass
plt.plot(friends_count)
plt.xlabel("Tweet #")
plt.ylabel("Friends")
figure(num=None, figsize=(8,6), dpi=80, facecolor='w', edgecolor='k')
plt.show()

for user in Users:
    try:
        friends_count.append(user['statuses_count'])
    except:
        pass
plt.plot(friends_count)
plt.xlabel("Tweet #")
plt.ylabel("Statuses")
figure(num=None, figsize=(8,6), dpi=80, facecolor='w', edgecolor='k')
plt.show()

for user in Users:
    try:
        friends_count.append(user['favourites_count'])
    except:
        pass
plt.plot(friends_count)
plt.xlabel("Tweet #")
plt.ylabel("Favourites")
figure(num=None, figsize=(8,6), dpi=80, facecolor='w', edgecolor='k')
plt.show()
```

```python
print(df['entities'][0])
Retweeted=df['retweeted']
rt_boolean_counts= Counter(Retweeted)
rt_hist = pd.DataFrame.from_dict(rt_boolean_counts, orient='index')
rt_hist.plot(kind='bar',legend=False)
plt.suptitle('Retweeted Status')
plt.show()
```

```python
# Origins of 10 most common tweets

city_list=[]
for user in Users:
    location= user['location']
    if location !="null":
        city_list.append(user['location'])
city_counts= dict(Counter(city_list).most_common(10))
df_hist = pd.DataFrame.from_dict(city_counts, orient='index')
df_hist.plot(kind='bar',legend=False)
plt.suptitle('Origin of Tweets')
plt.show()
```
